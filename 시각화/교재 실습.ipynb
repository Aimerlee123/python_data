```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('시도별 전출입 인구수.xlsx',header=0)
```

```python
df=df.fillna(method='ffill')
mask=(df['전출지별']=='서울특별시') & (df['전입지별']!='서울특별시')
df_seoul=df[mask]
```

```python
df_seoul=df_seoul.drop(['전출지별'],axis=1)
df_seoul.rename({'전입지별':'전입지'},axis=1,inplace=True)
df_seoul.set_index('전입지',inplace=True)
df_seoul
```

```python
sr_one=df_seoul.loc['경기도']
```

```python
from matplotlib import font_manager,rc
font_path="c:/Windows/Fonts/malgun.ttf"
font_name=font_manager.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)
plt.plot(sr_one.index,sr_one.values)
plt.title('서울->경기 인구 이동')

plt.xlabel('기간')
plt.ylabel('이동인구수')
plt.show()
```

```python
sr_one=df_seoul.loc['경기도']
plt.style.use('ggplot')
plt.figure(figsize=(14,5))
plt.xticks(rotation='vertical')
plt.plot(sr_one.index,sr_one.values)
plt.title('서울->경기 인구이동')
plt.xlabel('기간')
plt.ylabel('이동인구수')
```

```python
colyears=list(map(str,range(2010,2018)))
df2=df_seoul.loc[['충청남도','경상북도','강원도','전라남도'],colyears]
df2=df2.transpose()
df2
```

```python
df2.index=df2.index.map(int)
df2.plot(kind='bar',figsize=(20,10),width=0.7,color=['orange','green','skyblue','blue'])
plt.title('서울->타시도 이동인구',size=30)
plt.ylabel('이동인구수',size=20)
plt.xlabel('기간',size=20)
plt.ylim(5000,30000)
plt.legend(loc='best',fontsize=15)
```

```python
df3=df_seoul.loc[['충청남도','경상북도','강원도','전라남도'],colyears]
df3['합계']=df3.sum(axis=1)
dftotal=df3[['합계']].sort_values(by='합계',ascending=True)
```

```python
dftotal.plot(kind='barh')
plt.title('서울->타시도 인구이동')
plt.xlabel('이동인구수')
```
